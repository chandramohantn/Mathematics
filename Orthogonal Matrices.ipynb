```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def euclidean_norm(a):
    tot = 0
    for i in a:
        tot += i * i
    return np.around(np.sqrt(tot), decimals=3)
```

```python
b = np.array([3, 2])
```

```python
# Scenario 1 (With identity matrix)
a = np.array([[1, 0], [0, 1]])
style = dict(size=10, color='gray')

plt.subplot(1, 2, 1)
plt.plot([0, a[0, 0]], [0, a[0, 1]], 'b')
plt.plot([0, a[1, 0]], [0, a[1, 1]], 'b')
plt.plot([0, b[0]], [0, b[1]], 'r')
plt.text(1, 1, "Norm = " + str(euclidean_norm(b)), **style)

plt.subplot(1, 2, 2)
plt.plot([0, a[0, 0]], [0, a[0, 1]], 'b')
plt.plot([0, a[1, 0]], [0, a[1, 1]], 'b')
c = a.dot(b)
plt.plot([0, c[0]], [0, c[1]], 'g')
plt.text(1, 1, "Norm = " + str(euclidean_norm(c)), **style)
plt.tight_layout()
plt.show()
```

```python
# Scenario 2 (With orthogonal vectors but not orthonormal)
a = np.array([[1, 1], [1, -1]])
style = dict(size=10, color='gray')

plt.subplot(1, 2, 1)
plt.plot([0, a[0, 0]], [0, a[0, 1]], 'b')
plt.plot([0, a[1, 0]], [0, a[1, 1]], 'b')
plt.plot([0, b[0]], [0, b[1]], 'r')
plt.text(2, 1, "Norm = " + str(euclidean_norm(b)), **style)

plt.subplot(1, 2, 2)
plt.plot([0, a[0, 0]], [0, a[0, 1]], 'b')
plt.plot([0, a[1, 0]], [0, a[1, 1]], 'b')
c = a.dot(b)
plt.plot([0, c[0]], [0, c[1]], 'g')
plt.text(2, 1, "Norm = " + str(euclidean_norm(c)), **style)
plt.show()
```

```python
# Scenario 3 (With orthonormal vectors)
a = np.array([[1/np.sqrt(2), 1/np.sqrt(2)], [1/np.sqrt(2), -1/np.sqrt(2)]])
style = dict(size=10, color='gray')

plt.subplot(1, 2, 1)
plt.plot([0, a[0, 0]], [0, a[0, 1]], 'b')
plt.plot([0, a[1, 0]], [0, a[1, 1]], 'b')
plt.plot([0, b[0]], [0, b[1]], 'r')
plt.text(1, 1, "Norm = " + str(euclidean_norm(b)), **style)

plt.subplot(1, 2, 2)
plt.plot([0, a[0, 0]], [0, a[0, 1]], 'b')
plt.plot([0, a[1, 0]], [0, a[1, 1]], 'b')
c = a.dot(b)
plt.plot([0, c[0]], [0, c[1]], 'g')
plt.text(2, 0.5, "Norm = " + str(euclidean_norm(c)), **style)
plt.show()
```
