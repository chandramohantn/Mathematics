```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal
```

```python
freq = 1
sample_freq = 100
time = np.linspace(0, 10, 2 * sample_freq + 1)
amplitude = np.sin(2 * np.pi * freq * time)
plt.figure(figsize=(10,5))
plt.plot(time, amplitude)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
```

```python
freq = 1
sample_freq = 100
time = np.linspace(0, 10, 2 * sample_freq + 1)
square = np.sign(np.sin(2 * np.pi * freq * time))
plt.figure(figsize=(10,5))
plt.plot(time, square)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
```

```python
freq = 1
sample_freq = 100
time = np.linspace(0, 5, 2 * sample_freq + 1)
amplitude = np.sin(2 * np.pi * freq * time)
square = np.sign(np.sin(2 * np.pi * freq * time))
plt.figure(figsize=(20,10))
plt.plot(time, amplitude, color='r', label='Sine')
plt.plot(time, square, color='b', label='Square')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.legend(loc='upper right')
```

```python
time = np.linspace(0, 5, 2 * 100 + 1)
amplitude = np.sin(2 * np.pi * time)
square = np.sign(np.sin(2 * np.pi * time))
f = (4/np.pi) * np.sin(2*np.pi*time) + (4/(3*np.pi)) * np.sin(6*np.pi*time) + (4/(5*np.pi)) * np.sin(10*np.pi*time)
plt.figure(figsize=(20,10))
plt.plot(time, amplitude, color='r', label='Sine')
plt.plot(time, square, color='b', label='Square')
plt.plot(time, f, color='g', label='fourier (n=3)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.legend(loc='upper right')
```

```python
time = np.linspace(0, 5, 2 * 100 + 1)
amplitude = np.sin(2 * np.pi * time)
square = np.sign(np.sin(2 * np.pi * time))
f_10 = np.zeros((time.shape))
for n in range(1, 10, 2):
    f_10 += (4 / (n * np.pi)) * np.sin(2 * n * np.pi * time)
f_100 = np.zeros((time.shape))
for n in range(1, 100, 2):
    f_100 += (4 / (n * np.pi)) * np.sin(2 * n * np.pi * time)
#f = (4/np.pi) * np.sin(2*np.pi*time) + (4/(3*np.pi)) * np.sin(6*np.pi*time) + (4/(5*np.pi)) * np.sin(10*np.pi*time)
plt.figure(figsize=(20,10))
plt.plot(time, amplitude, color='r', label='Sine')
plt.plot(time, square, color='b', label='Square')
plt.plot(time, f_10, color='g', label='fourier (n=10)')
plt.plot(time, f_100, color='y', label='fourier (n=100)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True, which='both')
plt.legend(loc='upper right')
```
